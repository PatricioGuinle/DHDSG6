```python
import chardet
def get_encoding_type(csv_path):
    rawdata = open(csv_path, 'rb').read()
    result = chardet.detect(rawdata)
    return result.get('encoding')
```

```python
datos_ruta = "../Data/properati.csv"
datos_original = pd.read_csv(datos_ruta)
```

```python
#Creo una copia del DF
datos_temp = datos_original.copy(deep=True)
```

```python
#Columnas que son string
columnas_str = ['operation', 'property_type', 'place_name', 'place_with_parent_names', 'country_name', 'state_name', 'description', 'title']
```

```python
#Le quito los espacios en blanco
for columna in columnas_str:
    datos_temp[columna] = datos_temp[columna].str.strip()
```

```python
#BUSCAR NULOS
datos_temp["place_with_parent_names"].isnull().sum()
```

```python
#AGREGAR COLUMNAS DE UBICACION

def extraerPais(ubicacion):
    if ubicacion is not None:
        partes = ubicacion.split("|")
        
        if len(partes) >=2:
            return partes[1]
        else:
            return None

def extraerProvincia(ubicacion):
    if ubicacion is not None:
        partes = ubicacion.split("|")
        
        if len(partes) >=3:
            return partes[2]
        else:
            return None

def extraerDepartamento(ubicacion):
    if ubicacion is not None and \
    "Buenos Aires Costa Atlántica" not in ubicacion and \
    "Buenos Aires Interior" not in ubicacion and \
    "Capital Federal" not in ubicacion:
        partes = ubicacion.split("|")

        if len(partes) >= 4:
            return partes[3]
        else:
            return None
        
def extraerLocalidad(ubicacion):
    if ubicacion is not None:
        partes = ubicacion.split("|")

        if "Bs. As. Zona" in ubicacion and len(partes) >= 5:
            return partes[4]
        elif "Buenos Aires Costa Atlántica" in ubicacion and len(partes) >= 4:
            return partes[3]
        else:
            return None

def extraerCiudad(ubicacion):
    if ubicacion is not None:
        partes = ubicacion.split("|")

        if "Buenos Aires Interior" in ubicacion and len(partes) >= 4:
            return partes[3]
        else:
            return None
        
def extraerBarrio(ubicacion):
    if ubicacion is not None:
        partes = ubicacion.split("|")

        if "Capital Federal" in ubicacion and len(partes) >= 3:
            return partes[3]
        elif "Buenos Aires Costa Atlántica" in ubicacion and len(partes) >= 5:
            return partes[4]
        elif "Buenos Aires Interior" in ubicacion and len(partes) >= 5:
            return partes[4]
        else:
            return None

def extraerSubBarrio(ubicacion):
    if ubicacion is not None:
        partes = ubicacion.split("|")

        if "Palermo" in ubicacion and len(partes) >= 5:
            return partes[4]
        else:
            return None
        
datos_temp["Pais"] = datos_temp.place_with_parent_names.apply(extraerPais)
datos_temp["Provincia"] = datos_temp.place_with_parent_names.apply(extraerProvincia)
datos_temp["Provincia"] = datos_temp["Provincia"].replace(regex=[r'Buenos Aires Costa Atlántica'], value="Buenos Aires")
datos_temp["Provincia"] = datos_temp["Provincia"].replace(regex=[r'Buenos Aires Interior'], value="Buenos Aires")
datos_temp["Provincia"] = datos_temp["Provincia"].replace(regex=[r'Bs.As. G.B.A.*'], value="Buenos Aires")

datos_temp["Departamento"] = datos_temp.place_with_parent_names.apply(extraerDepartamento)
datos_temp["Localidad"] = datos_temp.place_with_parent_names.apply(extraerLocalidad)
datos_temp["Ciudad"] = datos_temp.place_with_parent_names.apply(extraerCiudad)
datos_temp["Barrio"] = datos_temp.place_with_parent_names.apply(extraerBarrio)
datos_temp["SubBarrio"] = datos_temp.place_with_parent_names.apply(extraerSubBarrio)

regexZonas = re.compile("Zona Norte|Zona Sur|Zona Oeste")
resultadoRegexZonas = datos_temp.place_with_parent_names.apply(lambda x: regexZonas.search(x))
datos_temp["Zona"] = resultadoRegexZonas.apply(lambda x: x if x is None else x.group(0))
```

```python
datos_temp.loc[:, ['Pais', 'Provincia', 'Zona', 'Departamento', 'Localidad', 'Ciudad', 'Barrio', 'SubBarrio']].to_csv('Ubicaciones.csv')
```
